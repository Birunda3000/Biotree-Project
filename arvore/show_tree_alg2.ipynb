```python
def buscar_esp(id_, tree_list):
    for esp in tree_list:
        if esp[4] == id_:
            return esp
def sort_func(item):
    return item[1]
def tree_ord(tree_list, origem_id, destino_id):           
    tree_list_ord = []
    tree_list.sort(key=sort_func)

    while len(tree_list) > 0:
        #--------------------------------------------------------
        pivo = tree_list.pop(0)
        pivo_id = pivo[4]
        pivo_sons = []
        aux_del = []#pq deleta o tree_list e diminui o tamanho logo salvo as deleções aqui

        if pivo_id in origem_id:

            for i in range(len(origem_id)):#procura o indice do pivo em origem

                if pivo_id == origem_id[i]:# quando acha

                    print('Conexão de', {origem_id[i]}, 'para', {destino_id[i]})
                    
                    if buscar_esp(destino_id[i], tree_list)!= None:# se tem 2 pais no primeiro pai ja tira na hora de buscar pelo segunda da erro
                        
                        pivo_sons.append( buscar_esp(destino_id[i], tree_list) )
                        
                        tree_list.remove(buscar_esp(destino_id[i], tree_list))
                    else:
                        print('Ja foi - ', {destino_id[i]})
                        
        pivo_sons.sort(key=sort_func, reverse=True)

        tree_list_ord.append(pivo)

        for son in pivo_sons:
            tree_list_ord.append(son)
        #--------------------------------------------------------
    return tree_list_ord
```

```python
def return_tree_data(df_especies, df_relations, ord_func):
    nome = df_especies['nome'].to_list()
    origem = df_especies['inicio'].to_list()
    extinção = df_especies['fim'].to_list()
    id_ = df_especies['id'].to_list()
    
    tree_especies_data = []
    for i in range(len(nome)):    
        tree_especies_data.append([nome[i], -origem[i], -extinção[i], i, id_[i]])
    for es in tree_especies_data:
        pass
        #print(es)
    
    O = df_relations['O-ID'].to_list()
    D = df_relations['D-ID'].to_list()
    
    #return tree_especies_data, O ,D # test
    
    tree_especies_data = ord_func(tree_especies_data, O, D)#ord
  
    x = []
    origem = []
    extinção = []
    nome = []
    id_ = []
    for i in range(len(tree_especies_data)):  
        nome.append(tree_especies_data[i][0])
        origem.append(tree_especies_data[i][1])
        extinção.append(tree_especies_data[i][2])
        x.append(i)
        id_.append(tree_especies_data[i][4])
#------------------------------------------------------------------------------------------------------
    x_o = []
    x_d = []
    y_l = []
    dy = []

    tree_r_data = []
    for i in range(len(O)):    
        tree_r_data.append([O[i], D[i]])
    for es in tree_r_data:

        ori = id_.index(es[0])
        des = id_.index(es[1])

        x_o.append(x[ori])
        x_d.append(x[des])

        dy.append(0)

        y_l.append(origem[des])
    
    return nome, origem, extinção, x, y_l, x_o, x_d, dy
```
